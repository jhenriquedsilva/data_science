```python
import requests
url = "https://api.covid19api.com/dayone/country/brazil"
answer = requests.get(url)
```

```python
raw_data = answer.json()
print(raw_data[-1])
```

```python
CONFIRMADOS = 0
OBITOS = 1
RECUPERADOS = 2
ATIVOS = 3
DATA = 4

for i in range(1, len(final_data)):
    final_data[i][DATA] = final_data[i][DATA][:10]
                    
for row in final_data:
    print(row)
```

```python
def get_datasets(data, label):
    if type(data[0] == list):
        datasets = []
        for i in range(len(data)):
            datasets.append({"labels": label[i], "data": data[i]})
        return datasets
    else:
        return [
            {
                "labels": label[0],
                "data": data
            }
        ]
```

```python
def set_title(title=""):
    if title != "":
        display = "true"
    else:
        display = "false"
    return {
        "title": title,
        "display": display
    }
```

```python
def create_chart(labels,data,label,kind="bar",title=""):
    datasets = get_datasets(data,label)
    options = set_title(title)
    
    chart = {
        "type": kind,
        "data":{"labels": labels, "datasets":datasets},
        "options": options
    }
    
    return chart
```

```python
def get_api_chart(chart):
    url_base = "https://quickchart.io/chart"
    answer = requests.get(f'{url_base}?c={str(chart)}')
    return answer.content
```

```python
def save_image(path, content):
    with open(path,"wb") as image: # Write a binary
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
y_data_1 = []
for obs in final_data[1::10]:
    y_data_1.append(obs[CONFIRMADOS])
    
y_data_2 = []
for obs in final_data[1::10]:
    y_data_2.append(obs[RECUPERADOS])
    
labels = ["Confirmados", "Recuperados"]

x = []
for obs in final_data[1::10]:
    x.append(obs[DATA].strftime("%d/%m/%Y")) # Datetime to string
    
chart = create_chart(x,[y_data_1,y_data_2],labels,title="GrÃ¡ficos confirmados vs recuperados")
chart_content = get_api_chart(chart)
save_image("meu_primeiro_grafico.png", chart_content)
display_image("meu_primeiro_grafico.png")
```

```python
from urllib.parse import quote
def get_api_qrcode(link):
    text = quote(link) # parsing do link para a url
    url_base = "https://quickchart.io/qr"
    answer = requests.get("{}?text={}".format(url_base,text))
    return answer.content
```
